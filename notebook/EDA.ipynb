### Tak away from Describe and info

3. Object colums are categories into 2 to 4 categories. We will convert categorical data into numerical one.

```python
data.nunique()
```

```python
# Encoding

#binary encoding

binary_cols = ["gender", "Partner", "Dependents", "PhoneService", "PaperlessBilling", "Churn"]

# mapping values to 0 or 1

data[binary_cols] = data[binary_cols].replace({
    'Yes':1, 'No':0, 'Male': 1, 'Female': 0 
}).astype(int)
data.head(3)
```

```python
service_col=["StreamingTV","StreamingMovies","InternetService","MultipleLines","PhoneService"]
data["n_services"] = (data[service_col] == "Yes").sum(axis=1)
```

```python
data.head(20)
```

```python
# Encoding
#Multi-categories encoding
from sklearn.preprocessing import LabelEncoder
multi_category_cols = ["MultipleLines", "InternetService", "OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport","StreamingTV","StreamingMovies","Contract","PaymentMethod"]

# mapping values to 0 or 1
label_encoder = LabelEncoder()

for col in multi_category_cols:
    data[col] = label_encoder.fit_transform(data[col])
    print(f"Mapping for column '{col}': {dict(zip(label_encoder.classes_, range(len(label_encoder.classes_))))}")
data.head(3)
```

```python
#convert all bool col to int
bool_cols_list = data.select_dtypes(include='bool').columns
data[bool_cols_list] = data[bool_cols_list].astype(int)
```

```python
#Totalcharges to numerical

data["TotalCharges"] = pd.to_numeric(data["TotalCharges"],errors='coerce')
data["TotalCharges"].dtype
```

```python
data["TotalCharges"].isnull().sum()
```
